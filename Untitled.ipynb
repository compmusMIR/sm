```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import librosa

y, sr = librosa.load('otherside.mp3')
chroma_cq = librosa.feature.chroma_cens(y=y, sr=sr)#, hop_length=512*16)
librosa.display.specshow(chroma_cq, sr=sr, y_axis='chroma', x_axis='time')
plt.title('chroma_cqt')
plt.colorbar()
plt.show()
```

```python
N = chroma_cq.shape[1]/32
chroma = np.ndarray((12,N))
for i in np.arange(N):
    chroma[:,i] = np.mean(chroma_cq[:,32*i:32*(i+1)],axis=1)

out = np.ndarray((N,N))
for i in np.arange(N):
    a = chroma_cq[:,32*i]
    for j in np.arange(N):
        b = chroma_cq[:,32*j]
        out[i,j] = 1-np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

print out[0:10,0]
```
