```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import librosa

y, sr = librosa.load('hardest.mp3')
C = np.abs(librosa.cqt(y, sr=sr,real=True))
librosa.display.specshow(librosa.logamplitude(C**2, ref_power=np.max),
                         sr=sr, x_axis='time', y_axis='cqt_note')

plt.colorbar(format='%+2.0f dB')
plt.title('Constant-Q power spectrum')
plt.tight_layout()
plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import librosa

y, sr = librosa.load('hardest.mp3')
chroma_cq = librosa.feature.chroma_cens(y=y, sr=sr)#, hop_length=512*16)
librosa.display.specshow(chroma_cq, sr=sr, y_axis='chroma', x_axis='time')
plt.title('chroma_cqt')
plt.colorbar()
plt.show()
```

```python
N = chroma_cq.shape[1]/2
    
out = np.ndarray((N,N))
for i in np.arange(N):
    a = chroma_cq[:,i]
    for j in np.arange(N):
        b = chroma_cq[:,j]
        out[i,j] = np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

print out[0:10,0]
```
