```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import librosa

y, sr = librosa.load('hardest.mp3')
C = np.abs(librosa.cqt(y, sr=sr,real=True))
librosa.display.specshow(librosa.logamplitude(C**2, ref_power=np.max),
                         sr=sr, x_axis='time', y_axis='cqt_note')

plt.colorbar(format='%+2.0f dB')
plt.title('Constant-Q power spectrum')
plt.tight_layout()
plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import librosa

y, sr = librosa.load('hardest.mp3',
                     offset=0, duration=180)
chroma_cq = librosa.feature.chroma_cqt(y=y, sr=sr, hop_length=128*2**6)
librosa.display.specshow(chroma_cq, y_axis='chroma', x_axis='time')
plt.title('chroma_cqt')
plt.colorbar()
plt.show()
```

```python
N = chroma_cq.shape[1]
out = np.ndarray((N,N))
for i in np.arange(N):
    a = chroma_cq[:,i]
    for j in np.arange(N):
        b = chroma_cq[:,j]
        out[i,j] = np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

print out[0:10,0]
```

```python
def display_image(ima, width=None, height=None):
    from io import BytesIO
    import PIL.Image
    from IPython.display import display, Image

    im = PIL.Image.fromarray(np.uint8(ima))
    bio = BytesIO()
    im.save(bio, format='png')

    if width is None and height is None:
        display(Image(bio.getvalue(),  format='png'))
    elif width is None:
        display(Image(bio.getvalue(), height=height,  format='png'))
    elif height is None:
        display(Image(bio.getvalue(), width=width,  format='png'))
    else:
        display(Image(bio.getvalue(), width=width, height=height,  format='png'))
        
display_image(np.rot90(out**10*255))
```
